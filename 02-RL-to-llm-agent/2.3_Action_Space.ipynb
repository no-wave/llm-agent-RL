# 2.3 행동 공간 (Action Space)

Agentic RL은 텍스트 + 도구 호출의 하이브리드 행동 공간을 사용한다.

- PBRFT: 텍스트만 생성
- Agentic RL: 텍스트 + 함수 호출 (하이브리드)

OpenAI API의 `tool_schema()`와 `run_tools()`를 사용한다.

```python
from utils_openai import *
```

## 실습 1: PBRFT의 텍스트 전용 행동

```python
heading("PBRFT: 텍스트 전용")

prompt = "2+2의 답과 설명을 해주세요다"
response = ask(prompt, temperature=0.5)

step_print(1, "프롬프트", prompt)
step_print(2, "응답 (텍스트만)", response[:100])
print("\
✓ PBRFT는 계산을 직접 하는 것으로 보인다.")
print("  하지만 실제로는 LLM의 학습 데이터에서 패턴을 회상할 뿐이다.")
```

## 실습 2: Agentic RL의 하이브리드 행동

```python
heading("Agentic RL: 하이브리드 행동")

calculator_tool = tool_schema(
    name="calculator",
    description="수학 계산을 수행한다",
    properties={"expression": {"type": "string", "description": "계산할 식다"}},
    required=["expression"]
)

tools = [calculator_tool]
messages = [
    {"role": "system", "content": "필요하면 계산 도구를 사용하세요다."},
    {"role": "user", "content": "2+2의 답을 계산 도구로 구하고 설명해주세요다"}
]

response = chat(messages=messages, tools=tools, temperature=0.5)
step_print(1, "요청", "계산 도구를 사용하여 2+2 계산")
step_print(2, "응답", str(response)[:150])
print("\
✓ Agentic RL은 도구를 활용하여 정확한 계산을 보장한다.")
```

## 핵심 비교

| 측면 | PBRFT | Agentic RL |
|------|-------|----------|
| **행동 공간** | 텍스트만 | 텍스트 + 도구 |
| **함수** | `ask()` | `chat()` + `tool_schema()` |
| **정확도** | 근사 | 높음 |
