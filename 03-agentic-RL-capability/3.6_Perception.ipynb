# 3.6 인식(Perception)


인식은 환경으로부터 정보를 감지하고 의미론적으로 해석하는 능력이다.

### 수동 vs 능동 인식

- **수동 인식**: 주어진 정보를 수동적으로 처리
- **능동 인식**: 필요한 정보를 주도적으로 요청



## 인식 파이프라인

### 인식 단계

```
관찰 → 해석 → 의미추출 → 저장
```

### 의미론적 해석

```
해석 = LLM(관찰 + 맥락)
```

### 능동 인식

에이전트가 필요한 추가 정보를 요청한다:
```
필요_정보 = LLM(현재_관찰 + 목표)
```


## 환경 설정

````python
from utils_openai import *

heading("3.6 인식(Perception)")
step_print(1, "API 준비", f"모델: {MODEL}")
````

## 실습 1: 관찰 해석

````python
step_print(1, "관찰 해석", "주어진 관찰을 의미론적으로 분석한다")
observation = "사용자가 빠른 응답을 원하고 정확성도 중요하다고 했다."
interpretation = ask(
    f"""다음 관찰을 분석하다:
{observation}

핵심 요구사항을 정리하다.""",
    temperature=0.5,
    max_tokens=100
)
print(f"분석: {interpretation}")
step_print(2, "중요도 평가", "관찰의 중요도를 판단한다")
importance = ask(
    f"""관찰: {observation}

0~10 중요도 점수만 반환하다.""",
    temperature=0.3,
    max_tokens=5
)
print(f"중요도: {importance}")
step_print(3, "정보 요청", "필요한 추가 정보를 파악한다")
additional_info = ask(
    f"""관찰: {observation}

더 알아야 할 것은 무엇인가?""",
    temperature=0.6,
    max_tokens=100
)
print(f"필요 정보: {additional_info}")
````
