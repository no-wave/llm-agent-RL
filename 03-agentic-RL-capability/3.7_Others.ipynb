3. **인간-AI 협업**: 사용자 피드백 활용



## 고급 주제들

### 신뢰도 점수

```
신뢰도 = 온도 역함수
낮은 온도 → 높은 신뢰도
높은 온도 → 낮은 신뢰도
```

### 불확실성 처리

- **높은 불확실성**: 사용자에게 확인 요청
- **낮은 불확실성**: 자동 결정

### 인간-AI 협업 루프

```
AI 제안 → 사용자 피드백 → 정책 업데이트 → 개선된 AI
```

## 실습 1: 신뢰도 평가

````python
step_print(1, "신뢰도 평가", "온도로 신뢰도를 계산한다")
low_temp_answer = ask("1 + 1 = ?", temperature=0.0, max_tokens=20)
high_temp_answer = ask("1 + 1 = ?", temperature=1.0, max_tokens=20)
print(f"낮은 온도 (높은 신뢰도): {low_temp_answer}")
print(f"높은 온도 (낮은 신뢰도): {high_temp_answer}")

step_print(2, "불확실성 처리", "신뢰도가 낮으면 사용자에게 확인한다")
confidence = 0.4  # 낮은 신뢰도

if confidence < 0.5:
    print("신뢰도가 낮아서 사용자에게 확인이 필요하다.")
    print("['옵션 A', '옵션 B', '둘 다 아님'] 중 선택해주세요.")
else:
    print("신뢰도가 높아서 직접 답변 가능하다.")

step_print(3, "피드백 활용", "사용자 피드백을 정책에 반영한다")
feedback = "답변이 너무 길었다. 더 간결하게 해줘."
improved_prompt = f"""이전 피드백: {feedback}

새로운 요청에 대해 간결하게 답변하다."""
print(f"개선된 프롬프트로 재생성하는 중...")
````
