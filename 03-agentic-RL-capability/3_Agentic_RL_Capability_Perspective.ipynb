# 3. 에이전트 RL: 역량 관점 (Agentic RL: The Capability Perspective)


에이전트 강화학습은 6가지 핵심 역량으로 구성된다:
1. **인식(Perception)**: 환경을 이해하다
2. **도구 사용(Tool Using)**: 문제 해결 도구를 선택하다
3. **메모리(Memory)**: 경험을 기억하고 상기하다
4. **계획(Planning)**: 목표 달성을 위한 행동을 계획하다
5. **추론(Reasoning)**: 논리적 결론을 도출하다
6. **자기 개선(Self-Improvement)**: 과거로부터 학습하다

이 6가지 역량은 **생성형 에이전트 메모리(Generative Agent Memory)**를 통해 통합된다:
- **인지 계층**: 인식으로부터 원시 관찰 → 의미론적 해석
- **회상 계층**: 메모리와 도구 사용으로부터 관련 경험 검색
- **성찰 계층**: 계획, 추론, 자기 개선으로부터 고수준 지식 생성

## 생성형 에이전트 메모리와 6가지 역량

### 핵심 개념

**생성형 에이전트 메모리** 구조:
```
관찰(Perception) → 인지 계층 → 저장
                    ↓
과거 경험 검색 → 회상 계층 → 맥락화
                    ↓
계획/추론 → 성찰 계층 → 고수준 지식
                    ↓
피드백 → 메모리 진화 → 다음 상호작용
```

### 6가지 역량의 역할

| 역량 | 기능 | 메모리 연결 |
|---|---|---|
| 인식 | 환경 관찰 의미 해석 | 인지 계층 |
| 도구 사용 | 문제 해결 도구 선택 | 회상 계층 |
| 메모리 | 정보 저장/검색 | 메모리 관리 |
| 계획 | 목표 달성 행동 계획 | 성찰 계층 |
| 추론 | 논리적 결론 도출 | 성찰 깊이 |
| 자기 개선 | 성과로부터 학습 | 메모리 진화 |

## 실습 1: 6가지 역량 개별 시연

각 역량을 간단한 예시로 보여준다.

````python
# 1. 인식(Perception) - 관찰을 해석하다
step_print(1, "인식 역량", "관찰을 의미론적으로 해석한다")

observation = "사용자가 복잡한 알고리즘 문제를 요청했다."

perception = ask(
    f"다음 관찰을 분석하고 핵심을 한국어 한 문장으로 요약하다: {observation}",
    system="당신은 관찰을 분석하는 지각 시스템이다.",
    temperature=0.5
)

print(f"  관찰: {observation}")
print(f"  분석: {perception}")
````

````python
# 2. 도구 사용(Tool Using) - 최적의 도구를 선택하다
step_print(2, "도구 사용 역량", "문제 해결을 위한 도구를 선택한다")

task = "웹에서 최신 Python 라이브러리 정보를 찾아야 한다."
available_tools = ["검색 엔진", "계산기", "문서 기반 검색", "코드 실행기"]

tool_selection = ask(f"""작업: {task}
이용 가능한 도구: {', '.join(available_tools)}

가장 적절한 도구 하나를 선택하고 이유를 한 문장으로 설명하라.""",
    system="당신은 문제 해결 도구를 선택하는 전문가다.",
    temperature=0.5
)

print(f"  작업: {task}")
print(f"  선택: {tool_selection}")
````

````python
# 3. 메모리(Memory) - 정보를 저장하고 검색하다
step_print(3, "메모리 역량", "경험을 저장하고 상기한다")

memory = MemoryStream()

# 메모리 저장
memory.add("Python은 데이터 과학에 많이 사용된다.", importance=0.8)
memory.add("사용자는 효율성을 중시한다.", importance=0.9)
memory.add("최근 머신러닝 프로젝트가 진행 중이다.", importance=0.7)

print(f"  저장된 메모리: {len(memory)} 개")

# 메모리 검색
retrieved = memory.retrieve("Python과 데이터", top_k=2)
print(f"  검색된 메모리: {len(retrieved)} 개")
for i, mem in enumerate(retrieved, 1):
    print(f"    {i}. {mem['content'][:40]}...")
````

````python
# 4. 계획(Planning) - 목표 달성 계획을 수립하다
step_print(4, "계획 역량", "목표 달성을 위한 행동을 계획한다")

goal = "Python으로 데이터 분석 프로젝트를 완성하다."
context = "팀은 3명이고 마감일이 2주 남았다."

plan = ask(
    f"""목표: {goal}
상황: {context}

단계별 계획을 3-4 단계로 한국어로 간결하게 제시하다.""",
    system="당신은 전략적 계획 수립 전문가다.",
    temperature=0.7,
    max_tokens=200
)

print(f"  목표: {goal}")
print(f"  계획:")
print(f"  {plan}")
````

````python
# 5. 추론(Reasoning) - 논리적 결론을 도출하다
step_print(5, "추론 역량", "관찰로부터 논리적 결론을 도출한다")

observation = "모든 머신러닝 모델은 훈련이 필요하고, 훈련에는 데이터가 필요하다."

reasoning = ask(
    f"""다음 관찰로부터 한국어로 2-3개의 논리적 결론을 도출하다:
{observation}",
    system="당신은 논리적 추론 전문가다.""",
    temperature=0.5,
    max_tokens=150
)

print(f"  관찰: {observation}")
print(f"  추론:")
print(f"  {reasoning}")
````

````python
# 6. 자기 개선(Self-Improvement) - 과거로부터 학습하다
step_print(6, "자기 개선 역량", "성과로부터 학습하여 개선한다")

past_experience = [
    {"action": "빠른 코딩", "outcome": "버그가 많았다"},
    {"action": "코드 리뷰", "outcome": "품질이 향상되었다"},
    {"action": "테스트 작성", "outcome": "신뢰도가 높아졌다"}
]

experience_text = "".join([f"- {p['action']}: {p['outcome']}" for p in past_experience])

improvement = ask(
    f"""다음 경험들로부터 배운 점을 정리하다:
{experience_text}

앞으로의 개선 방향을 한국어 2-3문장으로 제시하다.""",
    system="당신은 자기 개선 전문가다.",
    temperature=0.6,
    max_tokens=150
)

print(f"  과거 경험:")
for p in past_experience:
    print(f"    • {p['action']}: {p['outcome']}")
print(f" 개선 방향:")
print(f"  {improvement}")
````

## 실습 2: 생성형 에이전트 메모리 사이클

Perceive → Memory → Retrieve → Plan → Reflect 사이클을 실행한다.

````python
heading("생성형 에이전트 메모리 사이클")

# 메모리 스트림 생성
agent_memory = MemoryStream()

# Phase 1: Perceive - 관찰을 인지 계층에 저장
step_print(1, "인지(Perceive)", "관찰을 의미론적으로 해석하여 메모리에 저장한다")

observations = [
    ("사용자가 파이썬 프로그래밍을 배우고 싶어 한다.", 0.8),
    ("프로젝트 마감일이 한 주 남았다.", 0.9),
    ("팀 동료가 코드 리뷰를 완료했다.", 0.7)
]

for obs, importance in observations:
    agent_memory.add(obs, importance=importance)
    print(f"  ✓ {obs[:40]}... (중요도: {importance})")

print(f"메모리 저장됨: {len(agent_memory)} 개")
````

````python
# Phase 2: Retrieve - 회상 계층에서 관련 경험 검색
step_print(2, "회상(Retrieve)", "관련성 + 최신성 + 중요도로 메모리를 검색한다")

query = "파이썬 학습과 프로젝트 진행"
retrieved = agent_memory.retrieve(query, top_k=2)

print(f"  쿼리: {query}")
print(f"  검색된 메모리: {len(retrieved)} 개")
for i, mem in enumerate(retrieved, 1):
    print(f"    {i}. {mem['content']}")
````

````python
# Phase 3: Plan - 성찰 계층에서 계획 생성
step_print(3, "계획(Plan)", "검색된 메모리를 바탕으로 행동 계획을 수립한다")

retrieved_text = "\n".join([m['content'] for m in retrieved])

next_action = ask(
    f"""최근 상황:
{retrieved_text}

이를 바탕으로 이번 주의 행동 계획을 3 단계로 제시하다.",
    system="당신은 상황을 분석하고 계획을 세우는 전문가다.""",
    temperature=0.7,
    max_tokens=150
)

print(f"생성된 계획:")
print(f"  {next_action}")
````

````python
# Phase 4: Reflect - 성찰을 통해 통찰 생성 및 메모리 업데이트
step_print(4, "반성(Reflect)", "경험으로부터 높은 수준의 통찰을 도출한다")

insight = agent_memory.reflect(query)

print(f"도출된 통찰:")
print(f"  {insight}")
print(f"메모리에 반성이 저장됨 (의미적 메모리 강화)")
````

````python
# 최종 메모리 상태
print(f"최종 메모리 상태: {len(agent_memory)} 개")
````

## 실습 3: 역량 간 상호작용

6가지 역량이 생성형 에이전트 메모리를 통해 협력하는 방식을 보여준다.

````python
heading("역량 간 상호작용 시나리오")

# 시나리오: 복잡한 문제 해결
step_print(1, "인식", "문제를 이해한다")
problem = "팀이 점점 느려지는 시스템을 최적화해야 한다."
print(f"  문제: {problem}")

step_print(2, "메모리", "과거의 유사한 경험을 검색한다")
# 메모리에 최적화 관련 정보 추가
agent_memory.add("캐싱을 사용하면 성능이 10배 향상될 수 있다.", importance=0.8)
agent_memory.add("데이터베이스 인덱싱은 쿼리 속도를 크게 개선한다.", importance=0.7)
past_exp = agent_memory.retrieve("성능 최적화", top_k=2)
print(f"  찾은 경험: {len(past_exp)} 개")

step_print(3, "도구 선택", "최적의 해결책을 찾기 위한 도구를 선택한다")
tools = ask(
    f"""문제: {problem}
과거 경험: {past_exp[0]['content']}

이 문제를 해결하기 위해 사용할 수 있는 도구/기법을 2-3개 제시하다.""",
    temperature=0.6,
    max_tokens=100
)
print(f"  추천 도구: {tools[:100]}...")

step_print(4, "계획", "구체적인 실행 계획을 수립한다")
step_print(5, "추론", "결과를 예측한다")
step_print(6, "자기 개선", "실행 후 학습 사항을 메모리에 저장한다")

print("이러한 역량들이 생성형 에이전트 메모리를 통해 협력하여")
print("  복잡한 문제 해결을 가능하게 한다.")
````

1. **6가지 역량**: 에이전트는 인식, 도구 사용, 메모리, 계획, 추론, 자기 개선으로 구성된다.

2. **생성형 에이전트 메모리**: 3계층 메모리 아키텍처

3. **사이클**: Perceive → Store → Retrieve → Plan → Act → Reflect

4. **협력**: 각 역량이 메모리를 통해 상호작용하여 창발적 행동 생성
