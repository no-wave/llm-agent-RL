# 3.1 계획(Planning)

계획은 목표 달성을 위해 행동의 순서를 결정하는 역량이다.

### 두 가지 계획 방식

1. **휴리스틱 계획**: 고정된 규칙 사용 (빠르지만 유연하지 못함)
2. **학습된 메타-계획**: LLM이 맥락을 고려하여 계획 생성 (느리지만 적응적)

### System 1 vs System 2 사고

- **System 1**: 빠른 직관적 사고 (낮은 온도)
- **System 2**: 느린 숙고적 사고 (높은 온도, Chain-of-Thought)

## 휴리스틱 vs 학습된 메타-계획

### 휴리스틱 계획 (Heuristic Planning)

고정된 규칙을 따른다:
```
목표 → 고정 규칙 → 행동 순서
예: "큰 작업 → 분해 → 자료 수집 → 실행"
```

**장점**: 빠름, 예측 가능
**단점**: 맥락을 무시, 새로운 상황 대응 어려움

### 학습된 메타-계획 (Learned Meta-Planning)

LLM이 상황을 이해하고 맥락에 맞는 계획 생성:
```
목표 + 맥락 → LLM 추론 → 적응적 계획
```

**공식**:
```
Plan(goal, context) = LLM(goal + context)
  where temperature ∈ [0.3, 0.7]
```

**장점**: 맥락 인식, 창의적
**단점**: 느림, API 비용

## 실습 1: 휴리스틱 계획

````python
step_print(1, "휴리스틱 계획", "고정된 템플릿 사용")

def heuristic_plan(goal: str) -> list:
    """모든 상황에 동일한 계획을 반환한다."""
    return [
        "1. 목표 분석",
        "2. 필요 자원 확인",
        "3. 세부 작업 분해",
        "4. 실행",
        "5. 검토"
    ]

goal = "Python 데이터 분석 앱 개발"
plan = heuristic_plan(goal)

print(f"목표: {goal}")
print(f"휴리스틱 계획:")
for step in plan:
    print(f"  {step}")
````

## 실습 2: 학습된 메타-계획

````python
step_print(2, "메타-계획", "LLM이 맥락을 고려한 계획 생성")

goal = "Python 데이터 분석 앱 개발"
context = "팀: 2명 (초보자 1명), 기간: 1주"

meta_plan = ask(
    f"""목표: {goal}
제약: {context}

이 상황에 맞는 3-4단계 계획을 하다.""",
    system="당신은 현실적 계획 전문가다.",
    temperature=0.6,
    max_tokens=200
)

print(f"목표: {goal}")
print(f"제약: {context}")
print(f"메타-계획:{meta_plan}")
````

## 실습 3: System 1 vs System 2 사고

````python
step_print(3, "System 1 (빠른 직관)", "온도 0.3")

problem = "시스템이 느려졌다."

system1 = ask(
    f"문제: {problem} 즉시 취할 3 조치를 하다.",
    temperature=0.3,
    max_tokens=150
)

print(f"결과:{system1}")
````

````python
step_print(4, "System 2 (깊은 숙고)", "온도 0.7, Chain-of-Thought")

system2 = ask(
    f"""문제: {problem}

다음을 생각하다:
1. 가능한 원인들
2. 각각 검증 방법
3. 해결 계획

상세히 한국어로 설명하다.""",
    temperature=0.7,
    max_tokens=250
)

print(f"결과:{system2}")
````

## 핵심 개념

### 온도 선택 가이드

- **온도 ≤ 0.3**: 빠른 판단 (System 1)
- **온도 ∈ [0.4, 0.6]**: 균형잡힌 계획
- **온도 ≥ 0.7**: 깊은 숙고 (System 2)

### 계획 공식

```
적응적_계획 = LLM(목표 + 자원 + 제약)
```

맥락을 반영한 현실적 계획을 생성한다.
