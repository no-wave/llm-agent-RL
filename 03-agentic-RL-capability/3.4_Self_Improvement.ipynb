# 3.4 자기 개선(Self-Improvement)


자기 개선은 에이전트가 과거의 성공과 실패로부터 배워 스스로를 개선하는 능력이다.

### Generate→Evaluate→Improve 루프

```
생성(생각) → 평가(채점) → 개선(업데이트) → 반복
```



## 자기 개선 사이클

### 개선 공식

```
개선_정책 = 현재_정책 + LLM(과거_성과 + 피드백)
```

### 언어적 자기수정

1. **생성**: 에이전트가 응답을 생성한다
2. **평가**: LLM이 품질을 평가한다
3. **개선**: 피드백을 바탕으로 수정한다
4. **내재화**: 개선사항을 정책에 반영한다


## 환경 설정

````python
from utils_openai import *

heading("3.4 자기 개선(Self-Improvement)")
step_print(1, "API 준비", f"모델: {MODEL}")
````

## 실습 1: 생성 및 평가

````python
def generate_response(topic: str) -> str:
    return ask(f"주제: {topic}에 대해 짧게 설명하다.", temperature=0.7, max_tokens=100)
def evaluate_response(response: str) -> float:
    score = llm_reward(response, criteria="정확성, 완전성, 유용성", max_score=10.0)
    return score
def improve_response(original: str, score: float) -> str:
    if score < 5.0:
        return ask(f"""다음 응답이 점수 {score:.1f}를 받았다:
{original}

더 나은 버전으로 개선하다.""", temperature=0.8, max_tokens=100)
    return original

# 자기 개선 루프
topic = "강화학습의 기본 개념"
print(f"주제: {topic}")
for iteration in range(3):
    step_print(iteration + 1, f"반복 {iteration + 1}", "Generate → Evaluate → Improve")
    response = generate_response(topic)
    score = evaluate_response(response)
    print(f"생성된 응답: {response[:50]}...")
    print(f"점수: {score:.1f}")
    
    if iteration < 2:
        response = improve_response(response, score)
        print(f"개선됨")
````
