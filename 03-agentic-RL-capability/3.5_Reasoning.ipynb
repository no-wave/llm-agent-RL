# 3.5 추론(Reasoning)


추론은 관찰과 기억으로부터 논리적 결론을 도출하는 능력이다.

### 두 가지 추론 방식

1. **System 1**: 빠른 직관적 추론 (낮은 온도)
2. **System 2**: 느린 숙고적 추론 (높은 온도, Chain-of-Thought)



## 빠른 vs 느린 추론

### System 1 (빠른 추론)
- 온도: 0.0~0.3
- 시간: 빠름
- 특징: 결정적, 직관적

### System 2 (숙고적 추론)
- 온도: 0.7~1.0
- 시간: 느림
- 특징: Chain-of-Thought, 다단계 분석

### 메타-추론

```
추론_깊이 = LLM(문제_복잡도)
```

복잡한 문제는 System 2로 처리한다.

## 실습 1: System 1 vs 2

````python
step_print(1, "System 1 (빠른)", "온도 0.2 - 결정적")
problem = "2 + 3 * 4는 얼마인가?"
system1_answer = ask(problem, temperature=0.2, max_tokens=50)
print(f"답: {system1_answer}")

step_print(2, "System 2 (느린)", "온도 0.8 - Chain-of-Thought")
system2_answer = ask(
    f"""{problem}

단계별로 계산하다.""",
    temperature=0.8,
    max_tokens=150
)
print(f"답: {system2_answer}")
````
